```python
data = {
    'First Name': ['Alice', 'Bob', 'Charlie', 'David', 'Alice', 'Bob', 'Alice' , 'Alice' , 'Alice']*2,
    'Last Name' : ['Keza', 'Manzi', 'Twahirwa', 'Gasana', 'Keza', 'Manzi', 'Uwimana' , 'Uwimana' , 'Batamuriza']*2,
    'Subject': ['Math', 'Math', 'Science', 'Science', 'Math', 'Science', 'Math', 'Science', 'Math']*2,
    'Academic Year' : [2023, 2023, 2023, 2023, 2023, 2023, 2023, 2023,2023,2024, 2024, 2024, 2024, 2024, 2024, 2024, 2024, 2024],
    'Score': [85, 75, 90, 80, 95, 85, 60, 70, 85, 75, 90, 90, 73, 94, 90, 85, 60, 80]
}

df = pd.DataFrame(data)
```

```python
df
```

```python
pivoted = df.pivot_table(index = ['First Name', 'Last Name', 'Subject'], columns = 'Academic Year', values = 'Score')
```

```python
pivoted
```

```python
pivoted.columns
```

```python
pivoted.reset_index()
```

```python
pivoted.columns.name = None
```

```python
pivoted_new = pivoted.reset_index()
pivoted_new
```

```python
pivoted_new['Improvement'] = pivoted_new[2024] - pivoted_new[2023]
pivoted_new
```

```python
def classify_marks(Improvement):
    
    if Improvement < 0:
        return 'Decrease'
    elif Improvement == 0:
        return 'Stable'
    else:
        return 'Increase'
        
```

```python
pivoted_new['Position'] = pivoted_new['Improvement'].apply(classify_marks)
```

```python
pivoted_new
```

```python
pivoted_new[pivoted_new['Position'] == 'Decrease']
```

```python
melt_new = pivoted_new.melt(id_vars= ['First Name', 'Last Name', 'Subject'], value_vars=[2023, 2024])
```
